```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import kagglehub
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

### Dataset

```python
path = kagglehub.dataset_download("anubhabswain/brain-weight-in-humans")
```

```python
# List all files in the dataset folder
files = os.listdir(path)
print(files)
```

```python
df = pd.read_csv(os.path.join(path, "dataset.csv"))
df.head()
```

```python
X = df["Head Size(cm^3)"].values
X
```

```python
Y = df["Brain Weight(grams)"].values
Y
```

```python
# Feature data most likely to be in 2D array form, so convert the data into 2D from

X = X.reshape(-1,1)
```

### Train and Test data split (80% - 20%)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20)
```

### Model

```python
# call the model by creating object
lr = LinearRegression()

# train the model with our data
lr = lr.fit(X_train,Y_train)
lr
```

```python
# Predit the values by trained model with out testing data (X_test)

Y_predict  = lr.predict(X_test)
Y_predict
```

```python
# Check the R2 Score how much it is near to 1.

R2 = r2_score(Y_test, Y_predict)
```

```python
R2
```

```python
plt.scatter(X_test, Y_test, color='black', label = 'Actual')
plt.plot(X_test, Y_predict, color='red', label = 'Predicted')
plt.legend()
plt.grid(True, alpha=0.5)
```
